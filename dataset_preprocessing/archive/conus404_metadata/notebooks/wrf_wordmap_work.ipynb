```python
import pandas as pd
import xarray as xr
from importlib.machinery import SourceFileLoader
```

```python
# imports the module from the given path
metalib = SourceFileLoader("metadata_lib","../metadata_lib.py").load_module()
```

```python
work_dir = '/Volumes/USGS_NHM3/datasets/wrf-conus404/originals'

wrf_wordmap_file = '/Users/pnorton/Projects/National_Hydrology_Model/FY2022_projects/repos/planning/conus404/wrfout_wordmap.csv'
wrf_init_wordmap_file = 'test.csv'
wordmap_json_file = '../json/wrf_wordmap.json'
```

## Create initial wordmap file containing words (from the description attributes) and the count of occurrence

```python
# Read a wrfout file which is used to create the initial wordmap
df = xr.open_dataset(f'{work_dir}/wrfout_d01_2020-09-30_00:00:00', engine='netcdf4', decode_coords=False, chunks={})
```

```python
wrf_wordmap = metalib.init_wordmap(df)
```

```python
word_df = pd.DataFrame(wrf_wordmap, index=[0]).transpose()
word_df.index.name = 'word_orig'
word_df.rename(columns={0: 'occurrence'}, inplace=True)
word_df.head()
word_df['word_new'] = ''
```

```python
# Write the wordmap to a CSV file
word_df.to_csv(wrf_init_wordmap_file, sep='\t')
```

## Create a wordmap json file from a CSV file

```python
word_map = metalib.read_wordmap(wrf_wordmap_file)
metalib.write_wordmap_json(wordmap_json_file, word_map)
```
